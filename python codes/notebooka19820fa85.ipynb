```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("/kaggle/input/engineering-colleges-in-india/engineering colleges in India.csv")
```

```python
dataset.head()
```

```python
dataset["overview"] = np.nan
dataset.head()
```

```python

for i in dataset.index:
#     dataset['overview'][i] = dataset.iloc['Campus Size[i]', 'Courses[i]', 'Facilities[i]', 'City[i]', 'State[i]', 'College Type[i]', 'Average Fees[i]'].to_string(header=False, index=False)
    
    temp = str(dataset['Campus Size'][i]) + str(dataset['Courses'][i]) + str(dataset['State'][i]) + str(dataset['College Type'][i]) + str(dataset['Average Fees'][i])
    dataset['overview'][i] = temp
```

```python
dataset.head(3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words="english")
dataset['overview'] = dataset['overview'].fillna("")
tfidf_matrix = tfidf.fit_transform(dataset['overview'])

```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
print(cosine_sim)
```

```python
indices_clg = pd.Series(dataset.index, dataset['College Name']).drop_duplicates()
indices_overview = pd.Series(dataset.index, dataset['overview']).drop_duplicates()
print(indices_overview[indices_clg['Netaji Subhas University of Technology']])

print(dataset['overview'][2])
```

```python
prefer = dataset['overview'][26]
print(prefer)
```

```python
def get_recommendation(preferences, cosine_sim = cosine_sim):
    ind = indices_overview[preferences]
    sim_scores = enumerate(cosine_sim[ind])
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    
    sim_index = [i[0] for i in sim_scores]
    print(dataset["College Name"].iloc[sim_index])
    
  
get_recommendation(prefer)
```
